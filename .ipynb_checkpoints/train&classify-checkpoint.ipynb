```python
def show_max(mess_img, show_max_number=True, threshold_abs=60, min_distance=10):
    lm_list = []
    fig1 = plt.figure(figsize=(20,20))
    ax1 = fig1.add_subplot(111, aspect='equal')
    plt.imshow(mess_img, cmap="gray")
    for idx, loc_max in enumerate(peak_local_max(mess_img, threshold_abs=threshold_abs, min_distance=min_distance)):
        lm = Point(loc_max[0], loc_max[1])
        lm_list.append([lm])
        ax1.add_patch(patches.Circle((lm.y, lm.x), 5, fill=False, color="red"))
        if show_max_number:
            plt.text(lm.y, lm.x, str(idx), color="yellow")
    return lm_list
```

```python
def show_train_img(mess_img, curves, draw_circles=True):
    fig1 = plt.figure(figsize=(20,20))
    ax1 = fig1.add_subplot(111, aspect='equal')

    for idx, cr in enumerate(curves):
        for c in cr:
            mess_img[c.get_coord()] = 0

    plt.imshow(mess_img)
    if draw_circles:
        for idx, cb in enumerate(candidate_bubbles):
            ax1.add_patch(patches.Circle((cb.y, cb.x), cb.radius*1.4, fill=False, color="red"))
            plt.text(curves[idx][0].y, curves[idx][0].x, str(idx), color="yellow")
```
